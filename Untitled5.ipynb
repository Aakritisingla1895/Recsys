```python
import pandas as pd 
import numpy as np 
df1=pd.read_csv('ai_test.csv')
df2=pd.read_csv('rate.csv')
```

```python
df1.columns = ['author_name','affliations','article_title','doi', 'volume','type','userId' ]
df2= df2.merge(df1,on='userId')
```

```python
df2.head(15)
```

```python
C= df2['rating'].mean()
C
```

```python
m= df2['articleId'].quantile(0.9)
m
```

```python
q_articles = df2.copy().loc[df2['rating'] >= m]
q_articles.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['rating']
    R = x['articleId']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Define a new feature 'score' and calculate its value with `weighted_rating()`
q_articles['score'] = q_articles.apply(weighted_rating, axis=1)
```
