```python
import numpy as np # linear algebra
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
ratings=pd.read_csv('ai_cite.csv')
ratings.head(15)
```

```python
articles=pd.read_csv('ai_journal_articles_clean.csv')
articles.head(15)
```

```python
article_citations = pd.merge(articles, ratings)
article_citations.head(15)
```

```python
ratings_matrix = ratings.pivot_table(index=['user_id '],columns=['author_id'],values='cited').reset_index(drop=True)
ratings_matrix.fillna( 0, inplace = True )
ratings_matrix.head(15)
```

```python
movie_similarity=cosine_similarity(ratings_matrix)
np.fill_diagonal( movie_similarity, 0 ) 
movie_similarity
```
