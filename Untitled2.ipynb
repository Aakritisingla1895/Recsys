```python
import numpy as np # linear algebra
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
ratings=pd.read_csv('ai_cite.csv')
ratings.head(15)
```

```python
articles=pd.read_csv('ai_journal_articles_clean.csv')
articles.head(15)
```

```python
article_citations = pd.merge(articles, ratings)
article_citations.head(15)
```

```python
ratings_matrix = ratings.pivot_table(columns=['author_id'],values='cited').reset_index(drop=True)
ratings_matrix.fillna( 0, inplace = True )
ratings_matrix.head(15)
```

```python
article_similarity=cosine_similarity(ratings_matrix)
np.fill_diagonal( article_similarity, 0 ) 
article_similarity
```

```python
ratings_matrix = pd.DataFrame( article_similarity )
ratings_matrix.head(15)
```

```python
try:
    
    user_inp="artificial intellience"
    inp=articles[articles['title']==user_inp].index.tolist()
    inp=inp[0]
    
    articles['similarity'] = ratings_matrix.iloc[inp]
    articles.head(5)
    
except:
    print("Sorry, the article is not in the database!")
    
print("Recommended articles based on your choice of ",user_inp ,": \n", articles.sort_values( ["similarity"], ascending = False )[1:10])
```
